# Pruning missing values

```python
train_data['parentspecies'] = train_data['parentspecies'].fillna('toluene')
test_data['parentspecies'] = test_data['parentspecies'].fillna('toluene')

if 'ID' in train_data.columns:
    train_data_processed = train_data.drop(['ID'], axis=1)
else:
    train_data_processed = train_data

if 'ID' in test_data.columns:
    test_data_processed = test_data.drop(['ID'], axis=1)
else:
    test_data_processed = test_data

X = train_data_processed.drop('log_pSat_Pa', axis=1)  
y = train_data_processed['log_pSat_Pa'] 

print("Processed Train Data Shape:", X.shape)
print("Processed Test Data Shape:", test_data_processed.shape)
```

# Base Model

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

X_encoded = pd.get_dummies(X, columns=['parentspecies'], drop_first=True)
test_data_encoded = pd.get_dummies(test_data_processed, columns=['parentspecies'], drop_first=True)

X_encoded, test_data_encoded = X_encoded.align(test_data_encoded, join='left', axis=1, fill_value=0)

X_train, X_val, y_train, y_val = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_val)
r2 = r2_score(y_val, y_pred)
print("Validation R2 Score:", r2)
```

# CSV format

```python
test_original = pd.read_csv('test.csv')
test_ids = test_original['ID'] 

X_encoded = pd.get_dummies(X, columns=['parentspecies'], drop_first=True)
test_data_encoded = pd.get_dummies(test_data_processed, columns=['parentspecies'], drop_first=True)

X_encoded, test_data_encoded = X_encoded.align(test_data_encoded, join='left', axis=1, fill_value=0)

test_predictions = model.predict(test_data_encoded)

submission = pd.DataFrame({
    'ID': test_ids,  # Use IDs from the original test file
    'log_pSat_Pa': test_predictions
})

submission.to_csv('final_submission.csv', index=False)
print("Submission file 'final_submission.csv' created successfully!")
```
